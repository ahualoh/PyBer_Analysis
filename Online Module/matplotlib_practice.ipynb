```python
import matplotlib.pyplot as plt
%matplotlib inline
import statistics
import numpy as np
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot
plt.plot(x_axis, y_axis)
```

```python
stdev = statistics.stdev(y_axis)
stdev
```

```python
plt.errorbar(x_axis, y_axis, yerr=stdev,capsize=7)
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# ALTERNATE1: Create the plot with ax.plt()
fig = plt.figure()
ax = fig.add_subplot()
```

```python
# ^ above is empty, add line 4 below to display line
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
```

```python
# ALTERNATE2: Create the plot with ax.plt()
ax = plt.axes()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot.
plt.plot(x_axis, y_axis, marker="D", color="green", linewidth=2, label='Boston')

# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")

# Set the y limit between 0 and 45.
plt.ylim(0, 45)

# Create a title.
plt.title("PyBer Fare by Month")

# Add a grid.
plt.grid()

# Add the legend.
plt.legend()
```

## 5.1.5 
https://courses.bootcampspot.com/courses/2523/pages/5-dot-1-5-create-bar-charts-using-the-matlab-approach?module_item_id=806072

Reminder...
* x-axis = list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

* y-axis = list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]

```python
# Create the plot
plt.barh(x_axis, y_axis, color="magenta", label='Boston')
plt.legend(loc='best')
plt.title("PyBer Fare by Month")
plt.xlabel("Month")
plt.ylabel("Fare($)")
plt.gca().invert_yaxis()
```

```python
# Create the plot
plt.bar(x_axis, y_axis, color="gray", label='Boston',yerr=stdev, capsize=3)
plt.legend(loc='best')
plt.title("PyBer Fare by Month")
plt.xlabel("Month")
plt.ylabel("Fare($)")

```

```python
# Create the bar plot with ax.plt()
fig, ax = plt.subplots()
ax.bar(x_axis, y_axis, color="black", label='New York')
ax.legend(loc='best')
```

```python
# Create the hbar plot with ax.plt()
fig, ax = plt.subplots()
ax.barh(x_axis, y_axis, color="cyan", label='Chicago')
ax.legend(loc='best')
ax.set_title("Pyber Fare by Month")
ax.set_xlabel("Date")
ax.set_ylabel("Fare($)")
ax.invert_yaxis()
np.arange(0,51, step=5.0)
ax.set_xticks(np.arange(0, 51, step=5.0))
plt.show()
```
