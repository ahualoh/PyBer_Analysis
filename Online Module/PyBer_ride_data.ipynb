```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import statistics as stats
%matplotlib inline
```

```python
RideData_df = pd.read_csv("PyBer_ride_data.csv")
RideData_df
```

```python
RideData_df.plot(x="Month", y="Avg. Fare ($USD)")
plt.show()
```

```python
# Set x-axis and tick locations.
x_axis = np.arange(len(RideData_df))
tick_locations = [value for value in x_axis]

# Plot the data.
RideData_df.plot(x="Month", y="Avg. Fare ($USD)")
plt.xticks(tick_locations, RideData_df["Month"])
plt.show()
```

```python
stdev = stats.stdev(RideData_df["Avg. Fare ($USD)"])
stdev
```

```python
# Set y-axis and tick locations.
RideData_df.plot(kind='bar',
                 x="Month", 
                 y="Avg. Fare ($USD)",
                 yerr=stdev,
                 capsize=5,
                 color="skyblue")
plt.yticks(np.arange(0, max(RideData_df["Avg. Fare ($USD)"])+10,5))
plt.show()
```
